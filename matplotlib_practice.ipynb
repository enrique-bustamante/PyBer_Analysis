```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the same figure using ax.plot()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend
plt.plot(x_axis, y_axis, marker="D", color="green", linewidth=2, label='Boston')
# Create labels for the x and y axis
plt.xlabel('Date')
plt.ylabel('Fare ($)')
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Set a title
plt.title('PyBer Fare by Month')
# Add a grid
plt.grid()
# Add a legend
plt.legend()
```

```python
# Create the bar chart
plt.barh(x_axis, y_axis, color='magenta', label='Boston')
# Put the axis in descending order
plt.gca().invert_yaxis()
# Create labels for the x and y axes.
plt.xlabel('Date')
plt.ylabel('Fare ($)')
# Create a title
plt.title('PyBer Fare by Month')
# Add a legend
plt.legend()
```

```python
# Create the plot with ax.plot()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, color='k', label='New York')
```

```python
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color='cyan', label='Chicago')
ax.set_xlabel('Rate ($)')
ax.set_ylabel('Date')
ax.set_title('PyBer Rate by Month')
plt.gca().invert_yaxis()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

```

```python
# Create a scatterplot using MATLIB method
plt.plot(x_axis, y_axis, 'o')
```

```python

plt.scatter(x_axis, y_axis, s=[i*3 for i in y_axis])
```

```python
plt.scatter(y_axis, x_axis, color='red', label='Chicago')
plt.xlabel('Fare ($)')
plt.ylabel('Date')
plt.title('PyBer Fare per Month')
plt.gca().invert_yaxis()
```

```python
# Create a scatterplot with object-oriented interface
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s = [i * 5 for i in y_axis], color='skyblue',alpha=0.2, label='Boston',   edgecolor='black', ls='-', lw=2)
ax.set_title('PyBer Rates by Month')
ax.set_xlabel('Date')
ax.set_ylabel('Fare ($)')
ax.legend()
```

```python
# create a pie chart
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

plt.subplots(figsize=(8, 8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, 
    explode=explode_values, 
    labels=x_axis, 
    colors=colors, 
    autopct='%.1f%%')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 8))
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
ax.pie(y_axis, labels=x_axis, autopct='%.1f%%', shadow=True, colors=colors, explode=explode_values, startangle=-90)
plt.gca().invert_yaxis()
plt.show()
```
