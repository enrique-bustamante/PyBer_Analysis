```python
import matplotlib.pyplot as plt 
import numpy as np 
import pandas as pd 
import statistics as sts 
```

```python
pyber_ride_df = pd.read_csv('./Resources/Pyber_ride_data.csv')
```

```python
pyber_ride_df
```

```python
# Set x-axis and tick marks
x_axis = np.arange(len(pyber_ride_df))
tick_locations = [value for value in x_axis]
pyber_ride_df.plot(x = "Month", y = "Avg. Fare ($USD)")
plt.xticks(tick_locations, pyber_ride_df["Month"])
plt.show()
```

```python
# convert the code to display a bar chart
stdev = sts.stdev(pyber_ride_df['Avg. Fare ($USD)'])
x_axis = np.arange(len(pyber_ride_df))
tick_locations = [value for value in x_axis]
pyber_ride_df.plot(x = "Month", y = "Avg. Fare ($USD)", kind='bar', yerr=stdev, color='skyblue', capsize=3)
plt.xticks(tick_locations, pyber_ride_df["Month"], rotation=0)
plt.yticks(np.arange(0,51,5))
plt.show()
```
